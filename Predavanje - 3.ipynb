3. Učitaj broj igrača i onda njihova imena

```python
broj_igrača =input("Broj igrača: ")
while not broj_igrača.isnumeric():
    broj_igrača = input("Unesi broj igrača")
broj_igrača = int(broj_igrača)


broj_imena = 0
while broj_imena < broj_igrača:
    ime = input(f"Unesi ime: {broj_imena + 1}. člana")
    while True:
         if ime:
             break
         if ime.isalpha():
             break
         if ' ' in ime:
             break
         if ime.isnumeric():
             break
         ime = input(f"Unesi ime {broj_imena + 1}. član ")

    print(ime)
    broj_imena += 1


```

```python
import random
zamišljeni_broj = random.randint(1, 10)  # (1, 10 * TEŽINA) PRIMJER!

pokušaj = int(input('Unesi broj:'))

broj_pokušaja = 0
while pokušaj != zamišljeni_broj:
    if pokušaj < zamišljeni_broj:
        print('Zamišljeni broj je veči!')
    else:
        print('Zamišljeni broj je manji!')
    pokušaj = int(input('Unesi broj:'))


print('Čestitam!')
print(broj_pokušaja)

```
