Godine i mjesec

```python
god_rođ = int(input("Unesi godinu rođenja: "))
mj_rođ = int(input("Unesi mjesec rođenja: "))

godine = 2024 - god_rođ

if mj_rođ < 3:
    godine = godine + 1

print("Imaš", godine, "godina")
```

Paran/Neparan broj

```python
broj = int(input('Unesi broj: '))
if not broj % 2:
    print("Broj je paran!")
else:
    print('Broje je neparan!')
```

Prijestupna godina

```python
godina = int(input("Unesi godinu: "))

# prva logika: svake 4 godine
# 2024, 2028, 2032, ....
if godina % 4 == 0:
    print("L1: Godina", godina, "je prijestupna")

# druga logika: svake 4, ali ne svake 100
# 2024, 2060, 2080
# NE: 2100, 2200
if godina % 4 == 0:
    if godina % 100 != 0:
        print("L2: Godina", godina, "je prijestupna")

# treća logika:
# svake 4, ne svake 100, da svake 400
# DA: 2024, 2000
# NE: 2100
if godina % 4 == 0:
    if godina % 100 != 0 or godina % 400 == 0:
            print("L3: Godina", godina, "je prijestupna")
```
